Load the project libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Load & Inspect the project dataset

```python
df = pd.read_csv("Superstore.csv", encoding="latin-1")

df.head(5)
```

Remove Unnecessary Columns

```python
df.drop(columns = ["Row ID"], inplace = True)
```

More Data Inspection

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

Change data types of the Order Date & Ship Date columns

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

df.info()
```

Create Month column using the Order Date column

```python
df["Month"] = df["Order Date"].dt.to_period("M")

print(df["Month"])
```

Remove duplicates

```python
df.drop_duplicates()
```

Total Profit per Region (Column Chart)

```python
regional_profit = df.groupby("Region")["Profit"].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data= regional_profit, x="Region", y="Profit", color = "steelblue")
plt.title("Total Profit per Region")
plt.tight_layout()
plt.show()
```

Sales vs Profit per Month Year (Line Chart)

```python
month_sales_profit = df.groupby("Month")[["Sales","Profit"]].sum()
plt.figure(figsize=(20,10))
plt.plot(month_sales_profit.index.astype(str), month_sales_profit["Sales"], label = "Sales", color = "blue", marker = "o")
plt.plot(month_sales_profit.index.astype(str), month_sales_profit["Profit"], label = "Profit", color = "red", marker = "s")
plt.title("Monthly Sales vs Profit")
plt.xlabel("Month")
plt.ylabel("Amount ($)")
plt.xticks(rotation = 45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Top 10 Profitable Products (Bar Chart)

```python
product_profit = df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(15,10))
plt.barh(product_profit.index, product_profit.values, color = "steelblue")
plt.xlabel("Profit")
plt.ylabel("Product Name")
plt.title("Top 10 Profitable Products")
plt.gca().invert_yaxis()
plt.grid(axis="x")
plt.tight_layout()
plt.show()
```

Sales Distribution per Category

```python
category_distribution = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(5,5))
colors = sns.color_palette("deep")
plt.pie(category_distribution.values, labels = category_distribution.index, autopct="%1.1f%%", startangle=90, colors=colors)
plt.title("Sales Distribution per Category")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

Sales vs Profit (Interactive Scatterplot)

```python
fig = px.scatter(df, x="Sales", y="Profit", color = "Category",
                 hover_data=["Sub-Category","Region"],
                 title="Interactive Sales vs Profit by Category")
fig.show()
```
